
## Mini-projet : Calculatrice**

### **Niveau 1 : Calculette (basique)**  

**Fonctionnalités :**  
1. Effectuer des opérations mathématiques de base :  
   - Addition (+)  
   - Soustraction (-)  
   - Multiplication (*)  
   - Division (/)
2. Affichage du résultat  
3. Interface en mode console (ligne de commande)

**Contraintes :**  
- Gestion des erreurs de division par zéro  
- Saisie utilisateur sous la forme : `a opérateur b` (par exemple : `5 + 3`)  
- Validation des entrées pour éviter les erreurs de syntaxe (l'entrée est un nombre ?)  


### **Niveau 2 : Calculatrice collège**  

**Fonctionnalités :**  
1. Gestion des priorités opératoires 
   - Exemples : `2 + 3 * 4` ou `(5 + 3) * 2`.  
2. Prise en charge des parenthèses pour les expressions imbriquées
3. Interface en mode console améliorée (saisie d’une expression complète)

**Contraintes :**  
- Validation stricte des expressions saisies
- Gestion des erreurs de syntaxe et des parenthèses déséquilibrées


### **Niveau 3 : Calculatrice avec interface graphique**  
**Objectif :**  
Créer une calculatrice graphique intuitive pour les opérations de base  

**Fonctionnalités attendues :**  
1. Interface utilisateur graphique (GUI) :  
   - Boutons pour les chiffres et les opérations (+, -, *, /) 
   - Affichage des résultats dans une zone dédiée
2. Prise en charge des opérations de base (addition, soustraction, multiplication, division)  

**Contraintes :**  
- Interface ergonomique et adaptée à une utilisation au clavier et à la souris  
- Gestion des erreurs utilisateur (par exemple, bouton division par zéro désactivé) 

### **Niveau 4 : Calculatrice Lycée**  
**Objectif :**    

**Fonctionnalités attendues :**  
1. Gestion des fonctions mathématiques avancées :  
   - Trigonométrie (sin, cos, tan)
   - Logarithmes et exponentielles
   - Puissances et racines 
2. Traçage de courbes :  
   - Permettre la saisie d’une fonction (par exemple, `y = x^2 + 2x - 3`)
   - Affichage du graphique correspondant 
   - Navigation sur le graphique pour explorer les points clés (zoom et horizontal/vertical) 

**Contraintes :**  
- Interface graphique intuitive pour saisir les fonctions et afficher les résultats 
- Précision des graphiques avec axes adaptatifs  
- Gestion des erreurs (saisie invalide, division par zéro, etc)  

### **Niveau 5 : Calculatrice Lycée avancé**    

**Fonctionnalités attendues :**  
1. Prise en charge des fonctionnalités avancées :  
   - Calcul symbolique (dérivées, intégrales simples)
   - Résolution d’équations (par exemple, `2x + 3 = 7`) 
   - Analyse de limites et asymptotes
2. Traçage de courbes avancé :  
   - Affichage simultané de plusieurs fonctions  
   - Options pour dérivées, tangentes et intersections entre courbes  
3. Personnalisation des graphiques :  
   - Modification des couleurs, styles de lignes, et épaisseurs 
   - Exportation des graphiques en format image 

**Contraintes :**  
- Interface graphique avancée, mais intuitive (menu)
- Précision dans les calculs symboliques
