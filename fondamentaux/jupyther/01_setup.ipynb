# Configuration de base

- Avoir un IDE comme vscode ou clion
- sous windows : installer MinGW
- ajouter le C:\MinGw\bin dans les path (variable d'environment -> variables systeme)
- avec le cmd faire `gcc (-o nom_fichier_apres_compilation ) nom_fichier_en_C`
- pour l'executer `./nom_fichier_apres_compilation(.exe)`

installation : https://www.youtube.com/watch?v=4IsGRamElQc

Exemple de commande : `gcc -o compile/01_setup  n-compile/01_setup.c`

# Exemples de code

`#include <stdio.h>` <- bibliotheque de base pour les entrée sorties (pour le print) <br>
`int main() {} ` <- definie la fonction principal avec son type de retour (int)<br>
`return 0;` <- stop le programme et renvoie 0


Exemple simple :

```C
#include <stdio.h>

// déclaration de la première fonction
void afficherMessage();

// déclaration de la deuxième fonction
int additionner(int a, int b);

int main() {
    // appel de la première fonction
    afficherMessage();
    
    // appel de la deuxième fonction et affichage du résultat
    int resultat = additionner(5, 3);
    printf("Le résultat de l'addition est : %d\n", resultat);

    return 0;
}

// première fonction (procédure plus exactement)
void afficherMessage() {
    printf("Bonjour, voici une fonction simple !\n");
}

// deuxième fonction
int additionner(int a, int b) {
    return a + b;
}
```	

Il faut declarer les fonctions au debut du programme pour le compilateur connaisse leur existence
