# Les Opérations en C

# Opérateurs arithmétiques
Pour effectuer des calculs simples

- **Addition (+)** : ajoute deux valeurs
- **Soustraction (-)** : soustrait une valeur à une autre
- **Multiplication (*)** : multiplie deux valeurs
- **Division (/)** : divise une valeur par une autre
- **Modulo (%)** : donne le reste d'une division entière

Exemple :
```c
#include <stdio.h>

int main() {
    int a = 10;
    int b = 3;
    printf("Addition : %d\n", a + b);
    printf("Soustraction : %d\n", a - b);
    printf("Multiplication : %d\n", a * b);
    printf("Division : %d\n", a / b);
    printf("Modulo : %d\n", a % b);
    return 0;
}
```

# Opérateurs relationnels
Ces opérateurs comparent deux valeurs et retournent un résultat booléen (vrai ou faux)

- **Égal à (==)**
- **Différent de (!=)**
- **Plus grand que (>)**
- **Plus petit que (<)**
- **Plus grand ou égal à (>=)**
- **Plus petit ou égal à (<=)**

Exemple :
```c
#include <stdio.h>

int main() {
    int a = 10, b = 20;
    if (a < b) {
        printf("a est plus petit que b\n");
    }
    return 0;
}
```

# Opérateurs logiques

- **ET logique (&&)** : retourne vrai si toutes les conditions sont vraies
- **OU logique (||)** : retourne vrai si au moins une condition est vraie
- **NON logique (!)** : inverse l'état logique d'une condition

Exemple :
```c
#include <stdio.h>

int main() {
    int a = 10, b = 20;
    if (a > 5 && b > 15) {
        printf("Les deux conditions sont vraies\n");
    }
    return 0;
}
```

# Opérateurs bit à bit

- **ET (&)**
- **OU (|)**
- **OU exclusif (^)**
- **Décalage à gauche (<<)**
- **Décalage à droite (>>)**

Exemple :
```c
#include <stdio.h>

int main() {
    int a = 5, b = 3; // 5 = 0101, 3 = 0011 en binaire
    printf("ET bit à bit : %d\n", a & b); // 0001
    printf("OU bit à bit : %d\n", a | b); // 0111
    return 0;
}
```

# Opérateurs d'affectation
Ces opérateurs servent à donner des valeurs aux variables

- **Truc de base (=)**
- **Avec opération (+=, -=, *=, /=, %=)**

Exemple :
```c
#include <stdio.h>

int main() {
    int a = 10;
    a += 5; // équivaut à a = a + 5
    printf("a = %d\n", a);
    return 0;
}
```
