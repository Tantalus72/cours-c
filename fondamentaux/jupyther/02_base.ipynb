# Fonction Main

La fonction `main` est le point central d'un programme.

Exemple : 

```c
#include <stdio.h>

int main() {
    printf("Bonjour, monde !\n");
    return 0;
}
```
`return 0;` : permet de savoir si le programme est arrivé à la fin sans erreur

# Inclusion des bibliothèques standard

Pour utiliser des fonctions  comme `printf`, on dois importer les bibliothèques appropriées en haut du fichier. 
```c
#include <nom_bibliothèques>
```

Boite à outils :

- `<stdio.h>` : fonctions pour les entrées/sorties standard comme :
  - `printf` pour afficher du texte
  - `scanf` pour lire des entrées utilisateur
  - `fopen`, `fclose` pour gérer des fichiers

- `<stdlib.h>` : Contient des fonctions utilitaires pour :
  - gestion dynamique de la mémoire comme `malloc` et `free`
  - conversion de chaînes en nombres avec `atoi` ou `atof`
  - arrêt du programme avec `exit`

- `<string.h>` : fonctions pour manipuler les chaînes de caractères comme :
  - `strcpy` pour copier des chaînes
  - `strlen` pour obtenir la longueur d’une chaîne
  - `strcmp` pour comparer des chaînes

- `<math.h>` : des fonctions math comme :
  - `sqrt` pour calculer la racine carrée
  - `pow` pour les puissances
  - `sin`, `cos`, `tan` pour les fonctions trigonométriques

# Structure typique d'un programme C

Voici une structure typique d'un programme C :

```c
#include <stdio.h>
#include <stdlib.h>

// Fonction principale
int main() {
    // Déclarations des variables
    int a = 10;
    int b = 20;
    int somme;

    // Calcul
    somme = a + b;

    // Affichage du résultat
    printf("La somme de %d et %d est %d\n", a, b, somme); // %d -> prend les valeurs dans la suite de l'instrucion 1 par 1 
    return 0;
}
```
